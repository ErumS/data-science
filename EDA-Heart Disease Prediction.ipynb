BOXPLOT for outlier detection and imputation

````python
categories = ['sex', 'chest_pain', 'blood_sugar', 'ecg', 'thal', 'major vessels']
````

````python
data.head()
````

````python
for col in data.columns:
  if (data[col].dtypes in [int, float]) and (col not in categories):
    boxplt = plt.boxplot(data[col])
    outliers = list([item.get_ydata() for item in boxplt['fliers']][0])
    max_cap = [item.get_ydata() for item in boxplt['caps']][1][0]
    min_cap = [item.get_ydata() for item in boxplt['caps']][0][0]
    if outliers:
      print("For {}, the caps are ({}, {}). Outliers are ".format(col, min_cap, max_cap), outliers)
      # print(data[data[col].isin(outliers)])
      cols = data[(data[col] <= max_cap) & (data[col] >= min_cap)]
      if col in ['cholestrol']:
        col_mean = cols[cols['sex'] == 'female'][col].mean()
      else:
        col_mean = cols[col].mean()

      combined_df = pd.concat([data, cols])

      result_df = combined_df.drop_duplicates(keep=False)
      data.loc[result_df.index, col] = col_mean
````

### Hypothesis Testing

People have heart disease if their cholestrol level is above 200 mg/dL.

H0 = No heart disease if cholestrol levels above 200

HA = heart disease if cholestrol levels above 200

````python
sns.kdeplot(
    data=data, x="cholestrol", hue="heart disease",
    cumulative=True, common_norm=False, common_grid=True,
)
plt.title('CDF of cholestrol with or without heart disease')
plt.show()
````

1. Two sample t-tests

Assumptions
Before conducting the two-sample t-test using Python let us discuss the assumptions of this parametric test. Basically, there are three assumptions that we can make regarding the data groups:

Whether the two samples data groups are independent. Whether the data elements in respective groups follow any normal distribution. Whether the given two samples have similar variances. This assumption is also known as the homogeneity assumption.

````python
from scipy.stats import ks_2samp
import scipy.stats as stats
````

````python
x = data[data['heart disease']==True]['cholestrol'].values
y = data[data['heart disease']==False]['cholestrol'].values
````

````python
print(f"Variance of cholestrol of people with heart disease : {np.var(x)}")
print(f"Variance of cholestrol of people without heart disease : {np.var(y)}")
````

````python
print(f"Here the ratio is {np.var(y)/np.var(x)}")
````

````python
ttest = stats.ttest_ind(a=x, b=y, equal_var=True)
ttest
if ttest[1]<0.05:
  print(f"Cholestrol levels are greater for people with heart disease compared to people without heart disease. We accept Alternate Hypothesis H1.")
else:
  print(f"Cholestrol levels are lesser for people with heart disease compared to people without heart disease. We Accept Null Hypothesis")
````

Find association of age with sex while having a heart disease

Null Hypothesis (H0): There is no association between sex and the presence of heart disease.

Alternative Hypothesis (H1): There is an association between sex and the presence of heart disease.

USE CHI SQUARE test

````python
from scipy.stats import chi2_contingency
````

````python
df_above_50 = data[data['age'] > 50]

# Create contingency table
contingency_table = pd.crosstab(df_above_50['sex'], df_above_50['heart disease'])

# Perform chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-square statistic:", chi2)
print("P-value:", p)
print("Degrees of Freedom:", dof)
print("Expected frequencies table:")
print(expected)
````

Since p value = 0.0002 is less than 0.05, we reject null hypothesis. So there is an association between sex and the presence of heart disease over the age of 50.
